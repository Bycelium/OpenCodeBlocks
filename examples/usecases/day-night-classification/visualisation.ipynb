```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# We can access the dataset through relative paths as such:
DAY_PATH = "./data/clear_images.npy"
NIGHT_PATH = "./data/night_images.npy"

import numpy as np

day_images = np.load(DAY_PATH)
night_images = np.load(NIGHT_PATH)

print("Day images shape:", day_images.shape)
print("Night images shape:", night_images.shape)
```

```python
# Each label has 5000 RGB images of size 224*224
# Let's plot some of them

plt.figure(figsize=(12, 4))
for i in range(5):
    plt.subplot(2, 5, i+1)
    plt.axis("off")
    plt.title("Day", fontsize=10)
    plt.imshow(day_images[i])
    plt.subplot(2, 5, i+6)
    plt.axis("off")
    plt.title("Night", fontsize=10)
    plt.imshow(night_images[i])
plt.show()
```
